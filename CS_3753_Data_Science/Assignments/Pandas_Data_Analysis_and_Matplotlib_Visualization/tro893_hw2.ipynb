```python
from google.colab import drive
drive.mount('/content/drive')

from matplotlib import pyplot as plt
import numpy as np
import pdb # this is for debugging your code
import pandas as pd

#Question 1
# Read the data from the file fdata.csv using read_csv() and save the data in df.
# The reference of read_csv(): https://www.geeksforgeeks.org/python-read-csv-using-pandas-read_csv/
# You need to set the parameters (sep, parse_dates, index_col) in read_csv() properly
#df = pd.read_csv(XXX, sep=XXX, parse_dates=XXX, index_col=XXX)

# Base path to the data files
basePath = "/content/drive/My Drive/Colab Notebooks/Data Science/Data/"

#1. your code is here

# read the data
df = pd.read_csv(basePath + 'fdata.csv', sep=',', parse_dates=['Date'], index_col='Date')

# plot the figure
df.plot()
plt.show()

#Question 2

# Save the data in the variable "data"
# Each row is saved in a tuple
#data = [('2016-10-03', 772.559998), ... ]

#2. your code is here
data = [
    ('2016-10-03', 772.559998),
    ('2016-10-04', 776.429993),
    ('2016-10-05', 776.469971),
    ('2016-10-06', 776.859985),
    ('2016-10-07', 775.080017)
]

# Convert the data to a DataFrame for plotting
df2 = pd.DataFrame(data, columns=['Date', 'Close'])
df2['Date'] = pd.to_datetime(df2['Date'])
df2.set_index('Date', inplace=True)

# Plot the data as a red line with round markers
plt.plot(df2.index, df2['Close'], 'ro-')

# Set the ticks, labels, titles, grids
plt.title('Alphabet Inc. Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')

# Turn on the minor ticks
plt.minorticks_on()

# Customize the major/minor grids
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')

# Show the plot
plt.show()

#Question 3
#3. your code is here
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

# plot bar chart
plt.bar(languages, popularity, color='blue')

# set the ticks, labels, titles, grids
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Turn on the grid
plt.grid(True)

# Show plot
plt.show()

#Question 4
#4. your code is here
# Plot the horizontal bar chart
plt.barh(languages, popularity, color='red')

# Set ticks, labels, titles, grids
plt.title('Popularity of Programming Languages')
plt.xlabel('Popularity')
plt.ylabel('Programming Languages')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Turn on the grid
plt.grid(True)

# Show the plot
plt.show()


#Question 5
weight1=[67,57.2,59.6,59.64,55.8,61.2,60.45,61,56.23,56]
height1=[101.7,197.6,98.3,125.1,113.7,157.7,136,148.9,125.3,114.9]
weight2=[61.9,64,62.1,64.2,62.3,65.4,62.4,61.4,62.5,63.6]
height2=[152.8,155.3,135.1,125.2,151.3,135,182.2,195.9,165.1,125.1]
weight3=[68.2,67.2,68.4,68.7,71,71.3,70.8,70,71.1,71.7]
height3=[165.8,170.9,192.8,135.4,161.4,136.1,167.1,235.1,181.1,177.3]

# your code is here to draw the scatter plot
plt.scatter(weight1, height1, color='blue', label='Group 1')
plt.scatter(weight2, height2, color='green', label='Group 2')
plt.scatter(weight3, height3, color='red', label='Group 3')

# Set ticks, labels, titles, grids
plt.title('Weight vs Height Scatter Plot')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
