```
# pick a random piece and a random legal move for that piece
def GetRandomMove(board, player):
  # Get all pieces with legal moves for the current player
  pieces_with_moves = GetPiecesWithLegalMoves(board, player)

  # If there are no pieces with legal moves, return None
  if not pieces_with_moves:
    return None, None

  # Select a random piece from the list of pieces with legal moves
  piece = random.choice(pieces_with_moves)

  # Get the list of legal moves for the selected piece
  legal_moves = GetListOfLegalMoves(board, piece)

  # Select a random move from the list of legal moves
  move = random.choice(legal_moves)

  return piece, move

```

One function will evaluate the board if a move is performed - give score for each of piece and calculate the score for the entire chess board.
```
# Instructions:
# this function will calculate the score on the board, if a move is performed
# give score for each of piece and calculate the score for the chess board
def evl(board, player):
  piece_values = {
    'p': 1, 't': 3, 'b': 3, 'r': 5, 'q': 9, 'k': 0,
    'P': 1, 'T': 3, 'B': 3, 'R': 5, 'Q': 9, 'K': 0,
  }
  score = 0
  for row in board:
    for piece in row:
      if piece != '.':
        score += piece_values[piece]
  return score if player == 'white' else -score


def minmax(board, depth, is_maximizing):
  if depth == 0 or IsCheckmate(board, 'white' if is_maximizing else 'black'):
    return evl(board, 'white' if is_maximizing else 'black')

  if is_maximizing:
    max_eval = float('-inf')
    for piece in GetPiecesWithLegalMoves(board, 'white'):
      moves = GetListOfLegalMoves(board, piece)
      for move in moves:
        new_board = [row[:] for row in board]
        MovePiece(new_board, piece, move)
        eval = minmax(new_board, depth - 1, False)
        max_eval = max(max_eval, eval)
    return max_eval
  else:
    min_eval = float('inf')
    for piece in GetPiecesWithLegalMoves(board, 'black'):
      moves = GetListOfLegalMoves(board, piece)
      for move in moves:
        new_board = [row[:] for row in board]
        MovePiece(new_board, piece, move)
        eval = minmax(new_board, depth - 1, True)
        min_eval = min(min_eval, eval)
    return min_eval


# Instructions:
# return the best move for the current player using the MinMax strategy
# to get the allocated points, searching should be 2-ply (one Max and one Min)
def GetMinMaxMove(board, curPlayer):
  best_move = None
  best_value = float('-inf') if curPlayer == 'white' else float('inf')

  pieces = GetPiecesWithLegalMoves(board, curPlayer)
  for piece in pieces:
    moves = GetListOfLegalMoves(board, piece)
    for move in moves:
      new_board = [row.copy() for row in board]
      MovePiece(new_board, piece, move)
      board_value = minmax(new_board, 1, curPlayer == 'black')
      if curPlayer == 'white' and board_value > best_value:
        best_value = board_value
        best_move = (piece, move)
      elif curPlayer == 'black' and board_value < best_value:
        best_value = board_value
        best_move = (piece, move)

  return best_move

```

```
def main():
  # Initialize and setup the board
  # print("Calling ChessBoardSetup...")
  board = ChessBoardSetup()
  player = 'white'
  turns = 0
  max_turns = 100


  # print("Board setup completed. Starting the game...")

  while not IsCheckmate(board, player) and turns < max_turns:
    clear_output(wait=True)
    DrawBoard(board)
    print(f"\n\nTurn {turns + 1}: {player.capitalize()}'s move")

    # Get move for current player
    if player == 'white':
      move_result = GetMinMaxMove(board, player)
    else:
      move_result = GetRandomMove(board, player)

    if move_result is None:
      print(f"{player.capitalize()} has no legal moves. Stalemate!")
      break

    from_pos, to_pos = move_result
    from_row, from_col = from_pos
    piece = board[from_row][from_col]
    print(f"{piece} to {to_pos}")


    if from_pos is not None and to_pos is not None:
      board = MovePiece(board, from_pos, to_pos)
    else:
      print(f"{player.capitalize()} has no legal moves. Stalemate!")
      break

    DrawBoard(board)
    time.sleep(0.5)

    # Switch player
    player = 'black' if player == 'white' else 'white'
    turns += 1

  # Check and print - Stalemate or Checkmate
  clear_output(wait=True)
  DrawBoard(board)
  if IsCheckmate(board, 'white'):
    print("Checkmate! Black wins.")
  elif IsCheckmate(board, 'black'):
    print("Checkmate! White wins.")
  else:
    print("Stalemate or maximum turns reached. Draw.")

# print("Starting the script...")
if __name__ == "__main__":
#  print("Executing main...")
  main()
```
